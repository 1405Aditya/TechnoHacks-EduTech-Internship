```python
Q1 = np.percentile(data['price'], 25, interpolation = 'midpoint')
Q3 = np.percentile(data['price'], 75, interpolation = 'midpoint')
```

```python
IQR = Q3 - Q1
```

```python
lowerbound = Q1 - 1.5 * IQR
upperbound = Q1 + 1.5 * IQR
```

```python
data = data[(data.price < upperbound) & (data.price > lowerbound)]
```

```python
x = data.drop(columns = 'price')
X = x.drop(columns = 'date')
y = data['price']
```

```python
X.head()
```

```python
y
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(X)
```

```python
X[:] = scaler.transform(X)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_train)
```

```python
y_pred[:5]
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso_model = Lasso(alpha = 50, max_iter = 100, tol=0.1) 
lasso_model.fit(X_train, y_train)                    
```

```python
lasso_pred = lasso_model.predict(X_test)
```

```python
lasso_pred
```

```python
mean_absolute_error(y_test,lasso_pred)
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha = 50, max_iter = 100, tol=0.1) 
ridge_model.fit(X_train, y_train) 
```

```python
ridge_pred = ridge_model.predict(X_test)
```

```python
ridge_pred
```

```python
mean_absolute_error(y_test,ridge_pred)
```
